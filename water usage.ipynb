```python
import requests
import pandas as pd
import matplotlib
import matplotlib.dates as mdates
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
matplotlib.rcParams["font.size"]=18
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv('Water_Usage_Data.csv')
df['date_applicable']= pd.to_datetime(df['date_applicable'])
df['day'] = df['date_applicable'].dt.day
df['month'] = df['date_applicable'].dt.month
df['year'] = df['date_applicable'].dt.year
df
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(20,10))
ax_list=ax
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#99ffce']
labels = ['Residential', 'Multi-Family', 'Commercial', 'Industrial','Public Authority']
explode = (0, 0.1, 0, 0,0)



df_2019 = df[df['year']==2019]
df_2019[['residential_gallons','multi_family_gallons','commercial_gallons','industrial_gallons','public_authority_gallons']].sum().plot.pie(startangle=100, ax = ax_list[0], title = "Water Usage in 2019", labels=labels, colors = colors, autopct='%1.1f%%', pctdistance=0.8, shadow =True, explode=explode)
ax[0].set_ylabel("")

df_2020 = df[df['year']==2020]
df_2020[['residential_gallons','multi_family_gallons','commercial_gallons','industrial_gallons','public_authority_gallons']].sum().plot.pie(startangle=70, ax = ax_list[1], title = "Water Usage in 2020", labels=labels, colors = colors, autopct='%1.1f%%', pctdistance=0.8, shadow =True, explode=explode)
ax[1].set_ylabel("")



# startangle=100, ax = ax, title = "Water Usage in 2019", labels=labels, colors = colors, explode = explode, autopct='%1.1f%%', pctdistance=0.8)
```
